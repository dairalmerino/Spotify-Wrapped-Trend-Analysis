## Research Question

Similarities in a song's danceabilty, valence, ad energy rating can be statistically significant when examined across different release years because these ratings demonstrate traits that set the foundation for why songs become popular. The higher ratings in any of these categories is indicative of a higher likelihood that the song will be popular and, if the ratings are all similarly high or low, it is likely to that the song can return to popularity or "comeback".

Therefore, we can use pairwise comparisons for each variable across different release years to understand how the numeric similarities of these ratings compare using adjusted p-values and confidence intervals to assist in our understanding of what could be statistically significant in how songs become popular enough to top the Billboard charts or key industry indicators like 2025 Spotify Wrapped list.

Every year in December 6 million Spotify users get a list called "Spotify Wrapped" which highlights the 100 most played songs by Spotify users during the year regardless of the song's release date. The struggling music industry invests in teams of analysts who use Spotify Wrapped data to understand music, cultural and data trends that could indicate the "next big comeback" of songs or artists from certain publishing years. These songs are measured using the Spotify audio feature 0.0–1.0 using the following categories: danceability (how easy it is to dance to), energy (intensity and activity during the song), and valence (musical positivity throughout the song). We will use these variables as well as the following variables: Release date (release_year), to answer the following question: 

In 2025, do the top 100 songs included in the 2025 Spotify Wrapped list share statistically signficant similarities in danceability, valence, and energy according to their release date which could indicate a potential "comeback"?

## Hypothesis


By analyzing and synthesizing the top 100 songs included in the 2025 Spotify Wrapped list, we believe the research will demonstrate that there are similarities between songs by release date such that directional trends will emerge.

From our analysis, we believe that there will be directional predictions for each audio feature. We predict that danceability will differ the most by release year due to songs which are popular, but may not be suited specifically for dancing. We also predict that the most consistent variable that will not differ too much by release year will be valence because upbeat and happy qualities of music is, unlike danceability, often listened to regardless of activity performed. The final variable, energy, will demonstrate moderate variability because this variable tends to be genre specific to high energy genres of music like techno or house music which are also subject to confounding variables like digital media platforms like TikTok which can play clips of music over and over in an upbeat, high energy video.

We also suspect that there will be certain confounding variables present across all of the variables such as the rise in popularity of TikTok which could potentially artifically positively influence the number of listens for certain artists present on the 2025 Spotify Wrapped list or account for outliers.

##  Dataset #1 - Spotify Wrapped 2025 | Top Songs & Artist Dataset

This dataset captures the top 200 songs and artists according to the 2025 Spotify Wrapped play list plus the all-time most streamed songs in Spotify history up to March 2026. The 2025 Spotify Wrapped list reflects the most popular streams, regardless of release date, that were streamed from January 1, 2025 to October 31, 2025 according to the [Spotify Wrapped measurement window](https://newsroom.spotify.com/2025-12-05/wrapped-methodology-explained/). In this dataset, the monthly listener counts for artists are a point in time up to March 2026 and all-time stream counts reflect totals January 1st, 2025 to March 2026.

It was created by [Syed Ali Taqi](https://www.kaggle.com/alitaqishah) who is a statistician at Comsats University in Pakistan. The dataset contains 200 entries and 14 variables (described below). 

In this project, the "Spotify Wrapped 2025 | Top Songs & Artist Dataset" serves as our base dataset for the exploration of our research question and hypothesis validation. We will examine danceability, valence, and energy by release year to understand if there are significant differences in these variables by release year.

### Metrics

- **song_title:** The name of the song
- **artist:** The main artist and any featured artists in the song
- **streams_2025_billions:** The total number of streams in 2025 (billions), January 1, 2025 to October 31, 2025
- **total_streams_billions:**	The all-time total streams (billions), up to March 2026
- **primary_genre:** The main genre that the music was classified under
- **bpm:** The beats per minute (tempo) of the song
- **duration_seconds:** The length of the song (seconds)
- **release_year:** The year the song was originally released
- **artist_country:** The country of origin for the primary artist
- **explicit:** Determination if the the song contains explicit content
- **danceability:** Rated using the Spotify audio feature 0.0–1.0 and how suitable the song is for dancing
- **energy:** Rated using the Spotify audio feature 0.0–1.0 and is a reference to the intensity and activity of the song
- **valence:** Rated using the Spotify audio feature 0.0–1.0 and represents the musical positivity of the song
- **acousticness:** Rated using the Spotify audio feature 0.0–1.0 and references that acoustic vs electronic composition of the song.
- **peak_global_chart_position:** The highest position reached on Spotify Global Chart up to March 2026

### Limitations and Concerns

This dataset primarily reflects Spotify Wrapped metrics and trends for the year 2025, which does not account for previous years even though the release dates extend beyond 2025. This limitation might demonstrate bias against newer songs with release dates before 2025 because the songs with earlier release dates have has more time to accumulate streams and poluarity than newly released songs. This bias would apply to all songs with a release date before or in 2020 representing 50% of the dataset (100 songs) and a five year popularity window where the relevant variables could be more readily normalized against 2025.

### Dataset#2 - TikTok Trending Tracks Dataset

This dataset was composed to analyze the popularity of trending TikTok tracks and understand the factors that influence a track's popularity. It was created by Team Dan (https://github.com/romeoben/DSC7-Sprint2-TeamDan), whose group participated in the Eskwelabs DSCN Sprint. The dataset contains 6,746 entries (3,744 after removing duplicates caused by songs appearing across multiple TikTok playlists) and 10 variables.

In this project, the TikTok dataset serves as a way to explore a potential confounding variable, TikTok virality, that may influence which songs appear on the 2025 Spotify Wrapped list. By comparing the audio features of TikTok trending songs to those of Spotify Wrapped songs, we can assess whether TikTok popularity may be influencing certain tracks' stream counts.

### Metrics

- **track_name**: The name of the song
- **artist_name**: The name the song's artist
- **release_date**: The date the track was released on Spotify, stored as a datetime value
- **danceability**: How suitable a track is for dancing (0-1 scale; higher = more danceable)
- **energy**: Perceptual measure of intensity and activity (0-1 scale; higher = more energetic)
- **valence**: Musical positiveness of a track (0-1 scale; higher = more positive/happy)
- **tempo**: Estimated beats per minute (BPM) of the track
- **loudness**: Overall loudness of the track in decibels (dB); typically ranges from -60 to 0
- **acousticness**: Confidence measure of whether a track is acoustic (0-1 scale; higher = more acoustic)
- **popularity**: Spotify popularity score (0-100; higher = more popular globally)

### Limitations and Concerns

This dataset primarily reflects TikTok trends in the Philippines and Southeast  Asia, which may not fully represent global TikTok trends. This geographic bias is evident in the 276 songs (7%) with a Spotify popularity score of 0, likely reflecting regional tracks that trended locally but did not gain global Spotify traction. Additionally, the dataset only covers songs up to 2021, creating a time gap with our 2025 Spotify Wrapped data. While TikTok trends frequently cross regional and temporal boundaries, these limitations should be considered when interpreting results. 

Despite these limitation, the dataset was retained because its audio features (danceability, energy, valence, acousticness) mirror the metrics used in our Spotify Wrapped analysis, making it the most suitable available dataset for exploring TikTok virality as a confounding variable. The zero-popularity songs were also retained as they represent a meaningful characteristic of the dataset rather than data errors.

#### Analysis of our Core Dataset: Spotify Wrapped 2025 | Top Songs & Artists 

Below is the analysis of our core dataset Spotify Wrapped 2025 | Top Songs & Artists using pairwise comparisons and histograms for analysis. In this analysis we will compare all three variables using pairwise comparisons. Then, we will analyze each variable (energy, danceability, valence) alongside peak chart position to be able to validate or invalidate our hypothesis that these variables are statistically significant in whether or not there will be a "comeback" or resurgence in popularity.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
spotifyData = pd.read_csv('data\\02-processed\\spotify_wrapped_2025_top50_songs_final.csv')
spotifyData.head()
```

```python
pairwise_spotifyData = sns.pairplot(spotifyData, vars = ['danceability', 'energy', 'valence'])
```

##### Spotify Wrapped Pairplot (danceability, energy, valence)

##### Histogram Observations:

* The diagonal histograms show the individual distribution of each variable.
* Danceability is concentrated between 0.6-0.85, which means that most top Spotify Wrapped songs are moderately danceable, while only a few songs have low danceability values
* The distribution is left-skewed.
* The histogram for energy is more uniformly spread across the range of 0.4-0.9, showing that energy is more variable among these top songs.
* Valence has clusters around 0.5-0.6 and 0.8-0.9, creating a bimodal distribution with two peaks. This suggests that many songs are either moderately positive or very positive, with fewer songs falling in between.

##### Scatter plot Observations:

* The danceability vs. energy scatter plot shows a weak positive trend. Songs with higher energy levels tend to also have higher danceability.
* The danceability vs. valence scatter plot does not show a strong linear relationship, suggesting that songs can be danceable regardless of whether they are emotionally positive or not.
*  Energy vs. valence also shows no strong correlation.
* Overall, the scatter plots suggest that the three variables are relatively independent of one another.

```python
spotifyDataHistogramDance = sns.histplot(data=spotifyData, x="peak_global_chart_position", y="danceability",  cbar=True)
spotifyDataHistogramDance.set(xlabel='peak global chart position')
spotifyDataHistogramDance.set_title('Danceability vs. Peak Global Chart Position')
```

#### Danceability vs. Peak Global Chart Position Analysis

The light-to-dark blue gradients indicate the density of songs at each combination, where darker shades represent a higher concentration of songs. Chart positions 1 through 5, with danceability values between 0.65-0.85, show the heaviest concentration. This suggests that the majority of songs ranked highest globally had moderate-to-high danceability values.

There are almost no songs with danceability values below 0.5 at any chart position, indicating that danceability is likely an important factor for songs that achieve global chart success.

There is also an outlier cluster around chart positions 20-22 with danceability values near 0.7. These songs charted well but at lower positions, possibly representing “comeback” songs from earlier release years. Their danceability still falls within the mid-to-high range, which is consistent with the rest of the top songs.

This suggests that when older songs regain popularity, they tend to share similar danceability characteristics with newly released successful songs.

```python
spotifyDataHistogramValence = sns.histplot(data=spotifyData, x="peak_global_chart_position", y="valence",  cbar=True)
spotifyDataHistogramValence.set(xlabel='peak global chart position')
spotifyDataHistogramValence.set_title('Valence vs. Peak Global Chart Position')
```

#### Valence vs. Peak Global Chart Position

Unlike danceability, valence values at chart positions 1 to 5 are spread widely from approximately 0.6 to 0.95. This broad spread suggests that emotional positivity is not as strong of a determining factor for global chart success as danceability.

Songs do not necessarily need to sound happy in order to reach the top of the charts. However, positions 1 to 3 show the darkest concentrations at valence values between 0.8-0.95, indicating that highly positive songs are still capable of achieving top chart positions.

Compared to danceability, there is greater tolerance for moderate valence levels among successful songs. There is also an outlier cluster near position 22 with lower valence values around 0.2-0.5. These may represent slower or more melancholic songs that charted later in the year or older releases that experienced a resurgence in popularity.

```python
spotifyDataHistogramEnergy = sns.histplot(data=spotifyData, x="peak_global_chart_position", y="energy",  cbar=True)
spotifyDataHistogramEnergy.set(xlabel='peak global chart position')
spotifyDataHistogramEnergy.set_title('Energy vs. Peak Global Chart Position')
```

#### Energy vs. Peak Global Chart Position

Songs at the highest chart positions, ranging from 1 to 5, form dense clusters between energy values of 0.65-0.85, with the darkest cells concentrated in that range. Unlike danceability, there are also lower-energy songs with values around 0.4-0.55 that appear near the top positions. This suggests that energy, compared to danceability, is more flexible in terms of chart success.

Compared with danceability, energy has a broader spread across positions 1 to 5, which aligns with the hypothesis that energy would show moderate variability. There are almost no songs with energy values below 0.4 at any chart position, suggesting that there is still a minimum energy threshold for chart success, although it is lower than the threshold for danceability.

The cluster at position 22 appears to have mid-range energy values and is generally consistent with the pattern observed in the previous two charts.
