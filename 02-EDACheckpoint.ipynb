```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tiktokData = pd.read_csv('data/02-processed/tiktok_final.csv')
tiktokData.head()
```

```python
tiktokPairwise = sns.pairplot(tiktokData,
                              vars = ['danceability', 'energy', 'valence']
)
plt.show()
```

```python
corr = tiktokData[['danceability', 'energy', 'valence', 'acousticness']].corr()
sns.heatmap(corr, annot = True, cmap = 'Blues')

plt.title('Correlation Heatmap of TikTok Song Features')
plt.show()
```

```python
sns.histplot(
    data = tiktokData,
    x = 'danceability',
    bins = 20,
    kde = True
)

plt.title('Distribution of Danceability in TikTok Songs')
plt.show()
```

```python
sns.scatterplot(
    data = tiktokData,
    x = 'energy',
    y = 'danceability',
    hue = 'valence'
)

plt.title('Energy vs Danceability')
plt.show()
```

```python
year_avg = tiktokData.groupby('year')[['danceability', 'energy', 'valence']].mean()

year_avg.plot(figsize = (10,5))

plt.title('Average TikTok Song Features by Year')
plt.ylabel('Average Value')
plt.show()
```
