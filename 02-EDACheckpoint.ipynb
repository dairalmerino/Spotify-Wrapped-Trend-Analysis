## Research Question

Similarities in a song's danceabilty, valence, ad energy rating can be statistically significant when examined across different release years because these ratings demonstrate traits that set the foundation for why songs become popular. The higher ratings in any of these categories is indicative of a higher likelihood that the song will be popular and, if the ratings are all similarly high or low, it is likely to that the song can return to popularity or "comeback".

Therefore, we can use pairwise comparisons for each variable across different release years to understand how the numeric similarities of these ratings compare using adjusted p-values and confidence intervals to assist in our understanding of what could be statistically significant in how songs become popular enough to top the Billboard charts or key industry indicators like 2025 Spotify Wrapped list.

Every year in December 6 million Spotify users get a list called "Spotify Wrapped" which highlights the 100 most played songs by Spotify users during the year regardless of the song's release date. The struggling music industry invests in teams of analysts who use Spotify Wrapped data to understand music, cultural and data trends that could indicate the "next big comeback" of songs or artists from certain publishing years. These songs are measured using the Spotify audio feature 0.0–1.0 using the following categories: danceability (how easy it is to dance to), energy (intensity and activity during the song), and valence (musical positivity throughout the song). We will use these variables as well as the following variables: Release date (release_year), to answer the following question: 

In 2025, do the top 100 songs included in the 2025 Spotify Wrapped list share statistically signficant similarities in danceability, valence, and energy according to their release date which could indicate a potential "comeback"?

## Background and Prior Work

> Spotify started as a small start-up in Stockholm, Sweden. They started as a response to the growing issues of piracy and intellectual property theft in the music industry (<a name="cite_note-6">source</a>). Their solution was to create a platform where artists could earn revenue from people streaming their songs <a name="cite_note-1">(source)</a> This was not without controversy and competition though, as Apple Music signed exclusive contracts with big name artists to get the chance to stream their music first <a name="cite_note-7">(source)</a>. In addition, many artists like Taylor Swift and Jay-Z pulled their music from the platform previously, claiming that Spotify was underpaying the artists, and that it was just an experiment(BBC) <a name="cite_note-4">(source)</a>. Ten years ago in 2016, Spotify Wrapped burst into the scene, providing users with data that summarized their past listening activity for a given year <a name="cite_note-3">(source)</a>. This marketing strategy profited off of the need for users to feel included and validated for their listening hours and music taste. Since the inception of Spotify Wrapped, they have been slightly tweaking a few of the "categories", categorizing people in a more subjective, seemingly random, way <a name="cite_note-2">(Petridis)</a>. 

> Spotify Wrapped has drawn an increasing amount of attention in the more recent years, and with this attention, came people challenging the validity and ethics of the sensation (<a name="cite_note-8">source</a>). Because Spotify Wrapped uses LLMs and algorithmic data, it can often be inaccurate or biased(<a name="cite_note-8">source</a>). Users frequently report that the algorithm favors shorter songs over longer ones, reporting frequency over total listening time (<a name="cite_note-8">source</a>). The biggest elephant in the room is the fact that it shows how much this large company is surveilling users, and how much data they can get off of someone <a name="cite_note-2">(Petridis)</a>. The shiny bow of Wrapped also hides the fact that many artists on the platform are still underpaid for their work<a name="cite_note-2">(Petridis)</a>. With the surge of AI in the world, it is also apparent that Spotify Wrapped and Spotify in general rely heavily on AI for insights into listening activity <a name="cite_note-2">(Petridis)</a>. 

> We are looking into this now because the impact of streaming services like Spotify is a double edged sword for artists. According to statistics, the physical sale of music has gone down, while streaming has boomed<a name="cite_note-5">(Kiger)</a>. The benefits of Spotify Wrapped for artists include providing them with statistics about their audience demographic, which albums are streamed the most, and in what contexts their music is listened in (Spotify Newsroom)<a name="cite_note-5">(Kiger)</a>. > Some previous examples of analysis, datasets and studies that have been conducted include "Wrap your head around it: algorithmic self-making and performances of taste on Spotify Wrapped" which provides data driven evidence that Spotify Wrapped lists are not random ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). If millions of people listen to the same songs despite having their personalized lists, those songs must share properties which make them so much appealing to a broader audience ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). We intend on analyzing whether the audio profiles of Spotify Wrapped Top 100 songs are actually similar based on standardized variables to each other than they would be by chance, and if this similarity varies by release years.

## Hypothesis


By analyzing and synthesizing the top 100 songs included in the 2025 Spotify Wrapped list, we believe the research will demonstrate that there are similarities between songs by release date such that directional trends will emerge.

From our analysis, we believe that there will be directional predictions for each audio feature. We predict that danceability will differ the most by release year due to songs which are popular, but may not be suited specifically for dancing. We also predict that the most consistent variable that will not differ too much by release year will be valence because upbeat and happy qualities of music is, unlike danceability, often listened to regardless of activity performed. The final variable, energy, will demonstrate moderate variability because this variable tends to be genre specific to high energy genres of music like techno or house music which are also subject to confounding variables like digital media platforms like TikTok which can play clips of music over and over in an upbeat, high energy video.

We also suspect that there will be certain confounding variables present across all of the variables such as the rise in popularity of TikTok which could potentially artifically positively influence the number of listens for certain artists present on the 2025 Spotify Wrapped list or account for outliers.
