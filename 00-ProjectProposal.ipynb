## Research Question

Every year in December 6 million Spotify users get a list called "Spotify Wrapped" which highlights the 100 most played songs by Spotify users during the year regardless of the song's release date. The struggling music industry invests in teams of analysts who use Spotify Wrapped data to understand music, cultural and data trends that could indicate the "next big comeback" of songs or artists from certain publishing years. These songs are measured using the Spotify audio feature 0.0–1.0 using the following categories: danceability (how easy it is to dance to), energy (intensity and activity during the song), and valence (musical positivity throughout the song). We will use these variables as well as the following variables: Release date (release_year), to answer the following question: 

In 2025, do the top 100 songs included in the 2025 Spotify Wrapped list share statistically signficant similarities in danceability, valence, and energy according to their release date which could indicate a potential "comeback"?


## Background and Prior Work


> Spotify started as a small start-up in Stockholm, Sweden. They started as a response to the growing issues of piracy and intellectual property theft in the music industry (<a name="cite_note-6">source</a>). Their solution was to create a platform where artists could earn revenue from people streaming their songs <a name="cite_note-1">(source)</a> This was not without controversy and competition though, as Apple Music signed exclusive contracts with big name artists to get the chance to stream their music first <a name="cite_note-7">(source)</a>. In addition, many artists like Taylor Swift and Jay-Z pulled their music from the platform previously, claiming that Spotify was underpaying the artists, and that it was just an experiment(BBC) <a name="cite_note-4">(source)</a>. Ten years ago in 2016, Spotify Wrapped burst into the scene, providing users with data that summarized their past listening activity for a given year <a name="cite_note-3">(source)</a>. This marketing strategy profited off of the need for users to feel included and validated for their listening hours and music taste. Since the inception of Spotify Wrapped, they have been slightly tweaking a few of the "categories", categorizing people in a more subjective, seemingly random, way <a name="cite_note-2">(Petridis)</a>. 

> Spotify Wrapped has drawn an increasing amount of attention in the more recent years, and with this attention, came people challenging the validity and ethics of the sensation (<a name="cite_note-8">source</a>). Because Spotify Wrapped uses LLMs and algorithmic data, it can often be inaccurate or biased(<a name="cite_note-8">source</a>). Users frequently report that the algorithm favors shorter songs over longer ones, reporting frequency over total listening time (<a name="cite_note-8">source</a>). The biggest elephant in the room is the fact that it shows how much this large company is surveilling users, and how much data they can get off of someone <a name="cite_note-2">(Petridis)</a>. The shiny bow of Wrapped also hides the fact that many artists on the platform are still underpaid for their work<a name="cite_note-2">(Petridis)</a>. With the surge of AI in the world, it is also apparent that Spotify Wrapped and Spotify in general rely heavily on AI for insights into listening activity <a name="cite_note-2">(Petridis)</a>. 

> We are looking into this now because the impact of streaming services like Spotify is a double edged sword for artists. According to statistics, the physical sale of music has gone down, while streaming has boomed<a name="cite_note-5">(Kiger)</a>. The benefits of Spotify Wrapped for artists include providing them with statistics about their audience demographic, which albums are streamed the most, and in what contexts their music is listened in (Spotify Newsroom)<a name="cite_note-5">(Kiger)</a>. Some downsides of Spotify's growth on the music industry, is the predatory practices they promote. Their new policies regarding what counds as a "stream" for artists largely only benefit large record labels and hurt small creators uploading for the first time <a name="cite_note-4">(source)</a>. The Michigan Economic journal found that Spotify's new streaming reveune rules only benefit large labels that exploit their artists<a name="cite_note-4">(source)</a>. A benefit though, of the vast variety that Spotify provides for its users, is that people are able to have more variation in their music taste. This might allow people to find smaller artists that they like, something that was less common with buying physical CDs <a name="cite_note-5">(Kiger)</a>. 

> Some previous examples of analysis, datasets and studies that have been conducted include "Wrap your head around it: algorithmic self-making and performances of taste on Spotify Wrapped" which provides data driven evidence that Spotify Wrapped lists are not random ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). Rather they are socially and algorithmically shaped and also have significant overlap across users’ top songs ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). The findings about the algorithm also suggest that the audio features of Spotify Wrapped songs may reflect what listeners strategically choose to listen to but it does not have to be something that they enjoy ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). If millions of people listen to the same songs despite having their personalized lists, those songs must share properties which make them so much appealing to a broader audience ( <a name="cite_note-8">(Valiati, V., Lupinacci, L., & Soares, 2026)</a>). We intend on analyzing whether the audio profiles of Spotify Wrapped Top 100 songs are actually similar based on standardized variables to each other than they would be by chance, and if this similarity varies by release years.

> In another study, "Spotify (Un)wrapped: how ordinary users critically reflect on Spotify's datafication of the self within creative workshops",  researchers examine user reflections on how Spotify turns personal behavior into data stories ("datafication of the self") <a name="cite_note-9">(Annabell & Rasmussen, 2024)</a>. This study is supported by  additional findings that Spotify’s own listener typologies are built from the core variables that we will be using in our project: danceability, energy, and valence (<a name="cite_note-10">source</a>). It also shows Spotify’s internal audio feature categories are connected to the behavioral and cultural processes which determine what is displayed in Wrapped (<a name="cite_note-10">source</a>). “Wrappification” gives an explanation for why the Wrapped 100 should show a non-random audio feature pattern in which the platform selects songs that fit in the listener template defined by features like danceability, valence, and energy (<a name="cite_note-10">source</a>). We intend to analyze patterns in audio features by their release years are not trends but factors of how Spotify structures the listener type, and how certain songs fit in these types and how some don't. 

> In the final study uncovered in our background research, "Spotify ‘Unwrapped’: An Exploration of Data-Based Value Generation on a Music Streaming Platform" (<a name="cite_note-10">Adenuga, 2022</a>), a music industry analyst investigates the value creation of Spotify Wrapped for Spotify, artists, and advertisers. In this study, Adenuga validates that the music industry already uses Spotify Wrapped data for market intelligence which makes it a highly valuable asset(<a name="cite_note-10">Adenuga, 2022</a>). Spotify uses large amounts of user data to personalize music, promote artists, and also help advertisers target the intended audience. Spotify creates value for many groups, for example, songs that again become popular might create additional money in the music industry (<a name="cite_note-10">Adenuga, 2022</a>). This value proposition directly supports the urgency and importance of our project.
